```python
import pandas as pd
```

```python
df=pd.read_csv('./deodorant_dataset.csv')
```

```python
from matplotlib import pyplot as plt
```

```python
df.head()
```

```python
df.columns
```

```python
df['Brand Name'].unique()
```

```python
# Get total sell of a brand
grouping_by_brand_name=df.groupby(['Brand Name'])
```

```python
data=grouping_by_brand_name['Sales in Last Month (Units)'].sum()
```

```python
brand_names=data.index
sales=data
```

```python
print(brand_names.shape)
print(sales.shape)
```

```python
plt.plot(brand_names,sales)
plt.xlabel('Brand Name')
plt.ylabel('Sales in last Month (Units)')
plt.title("Sales of Deodorant Brands")
plt.gcf().autofmt_xdate()
plt.show()
```

```python
discounts=grouping_by_brand_name['Price (INR)'].median()
```

```python
discounts
```

```python
sales=sales.map(lambda x: x/10000)
```

```python
sales
```

```python
plt.plot(brand_names,sales,label="Sales in last month in (10^4 units)")
plt.plot(brand_names,discounts,label='Median Price of Brand')
plt.xlabel('Brand Name')
plt.title("Deodorant Brands")
plt.gcf().autofmt_xdate()
plt.legend()
plt.show()
```

```python
# Styling the plot
plt.plot(brand_names,sales,label="Sales in last month in (10^4 units)",color='k',linestyle='--')
plt.plot(brand_names,discounts,label='Median Price of Brand',color='c',linestyle='-.')
plt.xlabel('Brand Name')
plt.title("Deodorant Brands")
plt.gcf().autofmt_xdate()
plt.legend()
# adding for padding; so graph would not cutoff
plt.tight_layout()
plt.show()
```

```python
# to add grid
plt.plot(brand_names,sales,label="Sales in last month in (10^4 units)")
plt.plot(brand_names,discounts,label='Median Price of Brand')
plt.xlabel('Brand Name')
plt.title("Deodorant Brands")
plt.gcf().autofmt_xdate()
plt.legend()
plt.grid(True)
plt.show()
```

```python
# using some built-in style
# i.e. fivethirtyeight, seaborn, ggplot
# it is not recommended that use color and linestyle with built-in style; because it has its own style set
# plt.style.use('fivethirtyeight')
plt.style.use('ggplot')
plt.plot(brand_names,sales,label="Sales in last month in (10^4 units)")
plt.plot(brand_names,discounts,label='Median Price of Brand')
plt.xlabel('Brand Name')
plt.title("Deodorant Brands")
plt.gcf().autofmt_xdate()
plt.legend()
plt.show()
```

```python
# Stack plots
import pandas as pd
```

```python
df1=pd.read_csv('./ReligionOfPeace.csv')
```

```python
df1.head()
```

```python
from matplotlib import pyplot as plt
```

```python
df1['Date']=pd.to_datetime(df1['Date'])
```

```python
df1
```

```python
# Extract x,y co-ordinated
group_by_date=df1.groupby(df1['Date'].dt.year)
killed=group_by_date['Killed'].sum()
injured=group_by_date['Injured'].sum()
```

```python
date=group_by_date.groups.keys()
```

```python
print(date)
print(killed)
print(injured)
```

```python
plt.stackplot(date,killed,injured,colors=['m','c'],labels=['Killed','Injured'])
plt.title("Radical Islamic Terrorism Attacks")
plt.legend(loc='upper right')
plt.show()
```
