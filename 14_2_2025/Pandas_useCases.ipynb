```python
# Cleaning Data: Casting datatype and handling missing values
```

```python
import numpy as np
```

```python
people_new={
    'first':['Corey','Jane','John','Chris',np.nan,None,'N/A'],
    'last':['Schafer','Doe','Doe','Schafer',np.nan,np.nan,'Missing'],
    'email':['@','@','@',None,np.nan,'@','N/A']
}
```

```python
df_new=pd.DataFrame(people_new)
```

```python
df_new
```

```python
df_new.dropna(axis=0,how='any')
```

```python
df_new.dropna(axis=0,how='any',subset=['email'])
```

```python
df_new.replace('N/A',np.nan,inplace=True)
df_new.replace('Missing',np.nan,inplace=True)
```
