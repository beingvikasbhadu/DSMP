```python
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
df=pd.read_csv("./deodorant_dataset.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
group_by_brand_name=df.groupby('Brand Name')
```

```python
sales=group_by_brand_name['Sales in Last Month (Units)'].sum()
price=group_by_brand_name['Price (INR)'].sum()
```

```python
brand_names=group_by_brand_name.groups.keys()
```

```python
print(type(sales))
```

```python
%matplotlib inline
```

```python
fig,ax=plt.subplots()
ax.plot(brand_names,sales)
plt.gcf().autofmt_xdate()
ax.set_title("DEODORANT BRANDS")
ax.set_xlabel("Brand Name")
ax.set_ylabel("Sales in last month")
plt.tight_layout()
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(10,12),sharex=False)
ax1.plot(brand_names,sales)
plt.gcf().autofmt_xdate()
ax1.set_title("DEODORANT BRANDS")
ax1.set_ylabel("Sales in last month")
ax2.plot(brand_names,price)
ax2.set_xlabel("Brand Names")
ax2.set_ylabel("Price")
# plt.tight_layout()
plt.show()
```
